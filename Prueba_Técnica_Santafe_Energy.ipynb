```python
# Importar las librerías que se van a utilizar
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
# %matplotlib inline--> sirve para mostrar los gráficos de Matplotlib directamente dentro de un notebook de Jupyter.
import seaborn as sns
```

## Crear un DataFrame

```python
# leer el archivo csv "GeneraciónCauya2024" (data de generación de la planta Cauya durante el 2024.)
df_energy=pd.read_csv('Datos_Prueba_Técnica_GeneraciónCauya2024.csv')
df_energy
```

```python
# Atributo shape (filas, columnas)
df_energy.shape
```

```python
# Tipo de data de cada col.
df_energy.dtypes
```

# 1. Datos faltantes 

```python
# numero de filas faltantes en cada columna (NaN)
df_energy.isnull().sum()
```

```python
print(f'Tamaño del dataset antes de eliminar valores nulos: {df_energy.shape}')
print('\n')
df_energy.dropna(inplace=True)
print(f'Tamaño del dataset después de eliminar los valores nulos: {df_energy.shape}')
print('\n')
df_energy.info()
```

```python
df_energy.columns
```

```python
# Eliminar col "Id" y "Values_code"
df_energy.drop(['Id','Values_code'], axis=1, inplace=True)
```

```python
df_energy.head()
```

```python
df_energy.tail(30)
```

```python
# Convertir fecha (Date) en formate datetime

df_energy['Date'] = pd.to_datetime(df_energy['Date'])  
```

```python
# Creación de col 'Generacion_diaria', es la suma de cada hora. 
df_energy['Generacion_diaria']= df_energy.iloc[:,1:25].sum(axis=1)
```

```python
df_energy.head()
```

```python
df_energy.tail(30)
```

```python
# Generación diaria de los últimos 30 días 
df_energy['Generacion_diaria'].tail(30).sum()
```

```python
# Crear una col con el mes al que le corresponde cada registro
df_energy['Mes'] = df_energy['Date'].dt.month
```

```python
df_energy
```

```python
df_agrupado = df_energy.groupby('Mes',sort=True)['Generacion_diaria'].sum().reset_index()
```

```python
df_agrupado
```

```python
# Gráfico de barras

plt.figure(figsize=(10,6))
sns.barplot(x=df_agrupado['Mes'], y=df_agrupado['Generacion_diaria'], palette='Blues')

# Etiquetas y título del gráfico
plt.xlabel('Mes', fontsize=12)
plt.ylabel('Generación de Energía', fontsize=12)
plt.title('Generación de Energía por Mes', fontsize=14)
plt.show()
```

## 4.	¿Cómo es la distribución y qué tan dispersa es la generación en cada mes?

```python
df_agrupado
```

```python
df_summary = df_energy.groupby('Mes')['Generacion_diaria'].agg(['mean','median', 'std', 'min', 'max'])
df_summary
```

```python
# Agrupación de cuantos registros hay por cada mes para verificar el promedio. 
df_energy.groupby('Mes')['Date'].nunique()
```

```python
# Tendencia de la generación de energía mensual (Gráfico de líneas)
plt.figure(figsize=(12,5))
sns.lineplot(x=df_energy['Mes'], y=df_energy['Generacion_diaria'], marker='o')
plt.xticks(rotation=0)
plt.title("Tendencia de la Generación de Energía Mensual")
plt.grid()
plt.show()
```

```python
# Diagrama de dispersión 
plt.scatter(df_agrupado['Mes'], df_agrupado['Generacion_diaria'],  color='blue')
plt.xlabel("Mes")
plt.ylabel("Generación Mensual")
plt.show()
```
